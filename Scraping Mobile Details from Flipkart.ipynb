```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
```

```python
response=requests.get("https://www.flipkart.com/search?q=mobiles&sid=tyy%2C4io&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_5_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_1_5_na_na_na&as-pos=1&as-type=RECENT&suggestionId=mobiles%7CMobiles&requestId=557da780-c554-4b5b-84ec-b2d7d7eb5792&as-backfill=on")
print(response)
```

```python
soup=BeautifulSoup(response.content,"html.parser")
print(soup)
```

```python
next_btn_link=soup.find("a",attrs={"class":"_3fVaIS"})
# print(next_btn_link.get("href"))
home_url="https://www.flipkart.com"
new_url=home_url+next_btn_link.get("href")
print(new_url)
```

```python
mobile_list = []
```

```python
all_mobiles = []
```

```python
cards=soup.find_all("div",attrs={"class":"_1UoZlX"})
count=1
print(f"{count}page***************")
while(True):
    for card in cards:
        mobile_list,i = [],0
        title_div=card.find("div",attrs={"class":"_3wU53n"})
        title_txt=title_div.text
        mobile_list.append(title_txt)
        
        detail_page_link=card.find("a",attrs={"class":"_31qSD5"})
        home_url="https://www.flipkart.com"
        mob_page_url=home_url+detail_page_link.get("href")

        response2=requests.get(mob_page_url)
        
        soup2=BeautifulSoup(response2.content,"html.parser")
        div_ul = soup2.find('div',attrs={'class':'_3WHvuP'})
        for tag in div_ul.find_all("li", recursive=True): 
            mobile_list.append(tag.text)
            i +=1
            if i>4:
                break

        rating_div=card.find("div",attrs={"class":"hGSR34"})
        rating_txt=rating_div.text
        mobile_list.append(rating_txt)
        
        price_div=card.find("div",attrs={"class":"_2rQ-NK"})
        price_txt=price_div.text
        mobile_list.append(price_txt)
        all_mobiles.append(mobile_list)
        
    next_btn_link=soup.find("a",attrs={"class":"_3fVaIS"})
    home_url="https://www.flipkart.com"
    new_url=home_url+next_btn_link.get("href")
    response=requests.get(new_url)
    soup=BeautifulSoup(response.content,"html.parser")
    cards=soup.find_all("div",attrs={"class":"_1UoZlX"})
    count+=1
    print(f"{count}page***************")
    if count>464:
        break
    
print("bye")
```
